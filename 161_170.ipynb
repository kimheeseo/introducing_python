```python
# 활용 161. 주어진 숫자를 천 단위로 구분하기
num=input('아무 숫자를 입력하세요.:')

if num.isdigit(): # 사용자가 입력한 문자열이 숫자로만 구성되어 있는 지 확인합니다.
  num=num[::-1]
  print('num값:',num)
  ret=''
  for i,c in enumerate(num):
    i+=1
    if i!= len(num) and i%3==0:
      ret+=(c+',')
    else:
      ret+=c
      print('ret값:',ret)
  ret=ret[::-1]
  print(ret)
else:
  print('입력한 내용 [%s]; 숫자가 아닙니다.' %num)
```

```python
# 활용 165. 스택 구현하기(append, pop)
mystack=[]

def putdata(data):
  global mystack
  mystack.append(data)
# putdata(data): data를 mystack의 요소로 추가합니다. mystack은 전역변수로 선언되어 있으므로
# 함수 내에서 global 키워드를 사용해 전역변수 mystack을 사용한다는 것을 명시.

def popdata():
  global mystack
  if len(mystack) ==0:
    return None
  return mystack.pop()
# mystack으로부터 데이터를 하나 추출하는 함수를 popdata()로 구현합니다.
# 리스트 객체의 pop()메소드는 리스트에서 마지막 요소를 추출하고 해당 요소를 리스트로부터 제거합니다.

putdata('데이터1')
putdata([3,4,5,6])
putdata(12345)
print('mystack값:',mystack)

ret=popdata()
print('\n')
while ret!=None:
  print('스택에서 추출할 데이터:',ret)
  print('pop한 값',mystack)
  print('\n')
  ret=popdata()
```

```python
# 활용 166, 문장에 나타나는 문자 빈도수 계산하기
def getTextFreq(filename):
  with open(filename,'r') as f:
    text=f.read()
    fa={}
    for c in text:
      if c in fa:
        fa[c]+=1
      else:
        fa[c]=1
    return fa

ret=getTextFreq('mydata.txt')
ret=sorted(ret.items(), key=lambda x:x[1],reverse=True)
# ret을 내림차순으로 정렬하여 ret에 재지정합니다. ret은 getTextFreq()가 리턴한 사전입니다.
# ret은 (문자, 빈도수)가 하나의 요소인 리스트 형식이므로, 문자, 빈도수를 ret으로부터 하나식 추출하여 화면에 보기 좋게 출력합니다.

for c,freq in ret:
  if c == '\n':
    continue
  print('[%c]->[%d]회 나타남.' %(c,freq))
```

```python
# 활용 169. 파일에서 특정 문자열 교체하기
t1=input('찾을 단어를 입력하세요.')
t2=input('변경할 단어를 입력하세요.')

with open('mydata.txt','r') as f:
  with open('mydata2.txt','w') as h: # 'mydata2.txt'를 텍스트 파일 쓰기 모드로 오픈하여 파일 핸들을 h로 둡니다.
    text=f.read()
    text=text.replace(t1,t2) # text에 등장하는 t1을 t2로 모두 변경합니다. 변경된 text를 'mydata2.txt'에 저장합니다.
    h.write(text)

print('[%s]를 [%s]로 변경하였습니다.' %(t1,t2))
```

```python
# 활용 170. URL에 접속하여 HTML 페이지 화면에 출력하기
from urllib.request import urlopen

url='https://www.python.org'
with urlopen(url) as f:
  doc=f.read().decode()
# f.read(): url로부터 해당 리소스 데이터를 바이트 스트림으로 읽어봅니다.
# html 소스 코드는 텍스트로 되어 있으므로, f.read().decode()로 바이트 스트림을 텍스트로 변환하고 doc으로 둡니다.
  print(doc)
```
