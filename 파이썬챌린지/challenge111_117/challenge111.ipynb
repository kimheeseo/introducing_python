**114) Books.csv 파일을 사용하며, 사용자에게 시작 년도와 끝 년도를 입력하라고 요청한다. 입력된 두 년도 사이에 출간된 모든 책을 출력하라.**

```python
import csv

start=int(input("enter a starting year:"))
end= int(input("enter an end year:"))

file=list(csv.reader(open("books.csv")))
tmp=[]

for row in file:
  tmp.append(row)
tmp=tmp[1:]
print('tmp값:\n',tmp)

for row in tmp:
  if int(row[2])>=start and int(row[2])<=end:
    print(row)
```

**116) books.csv 파일의 데이터를 리스트로 가져온다. 사용자에게 리스트를 표시하고, 리스트의 어떤 행을 삭제하고 싶은지를 선택하라고 요청하여 선택한 행을 삭제한다. 또한 수정하고 싶은 데이터를 선택하라고 요청하고 선택한 행을 수정한다. 데이터를 원래의 csv 파일에 다시 써서 기존 데이터를 수정된 데이터로 덮어쓴다.**

```python
import csv

file=list(csv.reader(open("books.csv")))
booklist=[]
for row in file:
  booklist.append(row)

for i in range(0,len(booklist)):
  print('booklist값:', booklist[i])
```

```python
x=0
for row in booklist:
  display=x, booklist[x]
  print(display)
  x=x+1
getrid=int(input("enter a row number to delete:"))
del booklist[getrid]
```

```python
x=0
for row in booklist:
  display=x, booklist[x]
  print(display)
  x=x+1
alter=int(input("enter a row number to alter:"))
```

```python
x=0
for row in booklist[alter]:
  display=x, booklist[alter][x]
  print(display)
  x=x+1
```

```python
part=int(input("which part do you want to change?"))
newdata=input("enter new data:")
booklist[alter][part]=newdata
print(booklist[alter])
```

```python
file=open("books.csv","w")
for row in booklist:
  newrecord=row[0]+","+row[1]+","+row[2]+"\n"
  file.write(newrecord)
file.close()
```

**117) 사용자의 이름을 묻고 두 개의 임의의 질문을 생성하는 간단한 수학 퀴즈를 만든다. 사용자의 이름과 질문, 답변 그리고 최종 점수를 csv 파일에 저장한다. 프로그램이 실행될 때마다 csv파일에 추가되며 덮어쓰지 않는다.**

```python
import csv
import random
score=0
name=input("what is your name:")
q1_num1=random.randint(10,50)
q1_num2=random.randint(10,50)
```

```python
question1=str(q1_num1)+"+"+str(q1_num2)+"="
ans1=int(input(question1))
realans1=q1_num1+q1_num2

if ans1==realans1:
  score = score +1
```

```python
q2_num1=random.randint(10,50)
q2_num2=random.randint(10,50)
question2=str(q2_num1)+"+"+str(q2_num2)+"="
ans2=int(input(question2))
realans2=q2_num1+q2_num2
```

```python
if ans2==realans2:
  score = score+1

file=open("QuizScore.csv","a")
newrecord=name+","+question1+str(ans1)+","+question2+str(ans2)+","+str(score)+"\n"
file.write(str(newrecord))

file.close()
```
