```python
# 중급  131. 사전 정렬하기(sorted)
names={'Mary':10999,'Sams':2111,'Aimy':9778,'Tom':20245,'Michale':27115,'Bob':5887,'Kelly':7855}
ret=sorted(names)
print('ret값:',ret)

def f1(x):
  return x[0]

def f2(x):
  return x[1]

ret2=sorted(names.items(),key=f1)
print('ret2값:',ret2)
# sorted()에 key 인자를 추가하면 사전을 다양한 형태로 정렬할 수 있습니다.
# sorted()의 key인자로 사용할 함수 f1을 정의합니다.
# f1이 sorted()의 key인자로 사용되면 정렬할 데이터의 첫 번째 요소를 기준으로 정렬합니다.

ret3=sorted(names.items(),key=f2)
print('ret3값:',ret3)

# 파이썬 내장함수 sorted(): 사전 자료를 인자로 입력받아 정렬할 수 있습니다.
# 사전의 각 요소는 키:값으로 되어 있습니다.
# sorted()함수에 사전을 인자로 입력하면 기본적으로 사전의 키를 오름차순으로 정렬한 결과를 리스트로 리턴합니다.
```

```python
# 중급 133. 코드값에 대응하는 문자 얻기(chr)
val=input('문자 코드값을 입력하세요')
val=int(val)
print('val값:',val) # int(): 정수형 자료로 변환합니다.

try:
  ch=chr(val) # chr()를 이용해 사용자가 입력한 숫자에 해당하는 문자를 ch에 대입하고 화면에 val, val의 16진수, ch를 각각 출력.
# chr()의 인자로 정수값을 입력하면 이 정수값에 해당하는 문자를 리턴합니다.
  print('ch값:',ch)
  print('코드값: %d[%s],문자:%s' %(val, hex(val),ch))
except ValueError:
  print('입력한 <%d>에 대한 문자가 존재하지 않습니다!' %val)
```

```python
# 중급 135. 이름없는 한줄짜리 함수 만들기(lambda)
add=lambda x,y:x+y
ret=add(1,3)
print('ret값:',ret)

funcs=[lambda x:x+'.pptx', lambda x:x+'.docx']
ret1=funcs[0]('intro')
print('ret1값:',ret1)

ret2=funcs[1]('Report')
print('ret2값;',ret2)
# lambda 함수: 함수 이름 없이 한 줄로 간단하게 정의되는 함수.
# lambda: 인자, 인자, ... : 실행코드

# funcs[0]: lambda x:x+'.pptx'이고, funcs[1]: lambda x: x+'.docx' -> 'intro.pptx', 'report.docx'

names={'Mary':10999,'Sams':2111, 'Aimy':9778, 'Tom':20245}
ret3=sorted(names.items(), key=lambda x:x[0])
print('ret3값:',ret3)
# 사전의 모든 키: 값 요소를 키를 기준으로 오름차순으로 정렬하는 예를 lambda 함수를 이용하면 더 단순해집니다.
```

```python
# 중급 136. 인자를 바꾸어 함수를 반복 호출하여 결과값 얻기(map)
f=lambda x:x*x
args=[1,2,3,4,5]
ret=map(f,args)
# args의 모든 요소를 f에 대입하여 결과를 얻은 후, map 객체로 리턴합니다.
# map 객체는 list()를 이용해 리스트로 변환할 수 있습니다.

print('list(ret)값:', list(ret))
```

```python
# 활용 138. 텍스트 파일을 한줄씩 읽고 출력하기(readline)
with open('/content/stockcode.txt','r',encoding='cp949') as f:
  line_num=1
  line=f.readline()

  while line: # line의 값이 빈 문자열일 때까지 while 루프를 반복합니다.
    print('%d %s' %(line_num, line),end='')
    print('line값:',line)
    line=f.readline()
    line_num += 1
  f.close()
```

```python
# 텍스트 파일의 용량이 매우 클 경우에는 read()로 한꺼번에
# 파일 내용을 읽어 들이는 것은 메모리문제를 야기할 수 있습니다.

# readline(): 텍스트 파일에서 한줄을 읽습니다.
# 한줄을 읽고 나면 파일을 읽기 시작하는 위치는 그 다음 줄의 맨 처음이 됩니다.
# readline(): 연속적으로 호출하여 파일의 끝에서 더 이상 읽을 내용이 없으면
# readline()은 빈 문자열을 리턴합니다.

# readline()을 연속적으로 호출하여 파일의 끝에서 더 이상 읽을 내용이 없으면 readline()은 빈 문자열을 리턴합니다.
# 한줄을 읽고 나면 파일을 읽기 시작하는 위치는 그 다음 줄의 맨 처음이 됩니다.
```

```python
# 활용 139. 텍스트 파일을 한줄씩 읽고 출력하기(readlines)
with open('/content/stockcode.txt','r',encoding='cp949') as f:
  lines=f.readlines()
  for lin_num, line in enumerate(lines):
    print('line_num+1 값:%d'%(line_num+1))
    print('line 값:%s' %line)
    print('%d %s' %(line_num+1, line))
  f.close()
```

```python
# 텍스트 파일을 한줄씩 읽는 방법: 파일 객체의 readlines()를 이용.
# readllines() : 텍스트 파일의 끝까지 한줄씩 읽어 각 줄을 요소로 하는 리스트로 리턴합니다.
# readlines()는 텍스트 파일의 모든 내용을 한꺼번에 읽기 때문에, 파일 크기가 매우 큰 경우
# 메모리 문제가 발생할 수 있으니 주의합니다.
```
